### 1. What is the distribution of bike prices?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Bike Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#The distribution of bike prices shows that most bike prices are clustered at the lower end, with a long tail towards
#the higher end.
```

### 2. How does the price vary by city?

```python
price_by_city = df.groupby('city')['price'].describe()
price_by_city.sort_values(by='mean', ascending=False).head(10)
```

### 3. What are the most common bike brands in the dataset?

```python
plt.figure(figsize=(10, 6))
top_brands = df['brand'].value_counts().nlargest(10)
colors = sns.color_palette("husl", len(top_brands))  # Using the "husl" palette for diverse colors
sns.barplot(x=top_brands.index, y=top_brands.values, hue=top_brands.index, palette=colors, dodge=False)
plt.title('Top 10 Most Common Bike Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend([], [], frameon=False)  # Removing the legend as it is not necessary
plt.show()
```

### 4. Which cities have the most used bikes listed for sale?

```python
plt.figure(figsize=(10, 6))
top_cities = df['city'].value_counts().nlargest(10)
colors = sns.color_palette("husl", len(top_cities))  # Using the "husl" palette for diverse colors
sns.barplot(x=top_cities.index, y=top_cities.values,hue = top_cities.index, palette=colors,dodge=False)
plt.title('Top 10 Cities with Most Used Bikes Listed for Sale')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend([],[],frameon=False)
plt.show()
```

### 5. Which brands have the highest average power?

```python
# Calculate average power by brand
average_power_by_brand = df.groupby('brand')['power'].mean().sort_values(ascending=False)

# Plot the top 10 brands with the highest average power
plt.figure(figsize=(12, 6))
average_power_by_brand.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Brands with Highest Average Power')
plt.xlabel('Brand')
plt.ylabel('Average Power (cc)')
plt.xticks(rotation=45)
plt.show()

average_power_by_brand.head(10)
```
